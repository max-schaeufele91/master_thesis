```python
import math

path_length = np.empty_like(trajectories, dtype=float)
print("PL: ", path_length.shape)
print("Ts: ", trajectories.shape)

count = 0

for trajectory in trajectories:
    distance = 0
    #print("T: ", trajectory.shape)
    #print("Tl: ", len(trajectory))
    #print(count)
    for i in range(len(trajectory)):
        #print("X: ", trajectory[i][0], " Y: ", trajectory[i][1], "i: ", i)
        if(i < len(trajectory)-1):
            distance += math.sqrt((trajectory[i][0] - trajectory[i+1][0])**2 + (trajectory[i][1] - trajectory[i+1][1])**2)
            #print("D: ", distance)

    path_length[count] = distance
    count += 1

print(path_length)
```

```python
print("Min: ", np.min(path_length))
print("Max: ", np.max(path_length))

b=100

plt.hist(path_length, bins=b)
plt.show()
```
